## 데이터 전처리, 데이터 분석 및 시각화

```python
import matplotlib.pyplot as plt

'''
가족과 함께 대화하는 시간이 많은 아동은 
친구들과 함께 보내는 시간이 적을까? (JCh17dsc031, JCh17dsc034)
'''
df = df_org[['JCh17dsc031', 'JCh17dsc034']]
df.columns =  ['가족','친구']
df = df.dropna(axis=0)

df['친구-가족'] = df['친구'] - df['가족']
df_grouped = df.groupby(by='가족').mean()


plt.plot(df_grouped.index, df_grouped['친구-가족'])
plt.xlabel('family')
plt.ylabel('friend - family')
plt.show()

df_grouped
```

**결론**

가족과 대화하는 시간이 많은 아동은
친구들과 함께 하는 시간이 적고,

가족과 대화하는 시간이 적은 아동은
친구들과 함께 하는 시간이 많은 것으로 보임.

물리적인 시간이 제한되어 있으므로 상대적으로 당연한 결과

## dsc039 욕설빈도 sfs37 자신에 대한 만족

```python
import matplotlib.pyplot as plt

df = df_org[['JCh17dsc039', 'JCh17sfs037']]
df.columns = ['욕설빈도', '자존감']
df = df.dropna(axis=0)

df_grouped = df.groupby(by='욕설빈도').mean()

plt.plot(df_grouped.index, df_grouped['자존감'])
plt.xlabel('freq')
plt.ylabel('self_confidence')
plt.show()

df_grouped
```
